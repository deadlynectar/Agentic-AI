```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite")
response = llm.invoke("Keyword search is better or semantic serch?")
print(response.text)
```

```python
response = llm.invoke([
    ["system", "You are helpful assistant that answers in short less than 5 line"],
    ["human", "Keyword search is better or semantic serch?"]
])
print(response.text)
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model="qwen/qwen3-32b", reasoning_format="parsed")
messages = [
    (
        "system",
        "You are helpful assistant that answers in short less than 5 line",
    ),
    (
        "human",
        "Keyword search is better or semantic serch?"
    ),
]
ai_msg = llm.invoke(messages)
print(ai_msg.content)
```
