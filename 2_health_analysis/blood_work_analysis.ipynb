```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()

llm = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite")
```

```python
with open("blood_work.txt", "r") as f:
    blood_report = f.read()

print(blood_report[:200])
```

```python
extraction_prompt = f"""
You are a medical data extraction assistant.

From the blood report below, extract ALL test values and classify each one as HIGH, LOW and NORMAL based on the reference ranges provided in the report.

Format your respose as:
- test name: value | Status: HIGH/LOW/NORMAL | Reference: range

Blood Report:
{blood_report}
"""

extraction_response = llm.invoke(extraction_prompt)
extracted_values = extraction_response.text # for model groq, it should be extraction_response.content, for model gemini-3.1-flash-lite, it should be extraction_response.text

print("=== Stage 1:Extracted Values ===")
print(extracted_values)
```

```python
diet_prompt = f"""
You are a clinical nutritionist specializing in Indian dietry habits.

Based on the blood work analysis below, write:
1. A short health summary in 4-5 lines explaining the patient's condition in simple language.
2. A short, practical Indian diet plan having only two sections (1) Foodds to avoid (2) Foods to eat more of.
 Do not include any other section in diet plan.

 Blood work analysis:
    {extracted_values}
"""
diet_response = llm.invoke(diet_prompt)
diet_plan = diet_response.text # for model groq, it should be diet_response.content, for model gemini-3.1-flash-lite, it should be diet_response.text

print("=== Stage 2: Diet Plan ===")
print(diet_plan)
```
